```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

file_path = '/content/sample_data/netflix_titles.csv.csv'

df = pd.read_csv(file_path)
df.head()
# Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')

# Create new columns
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Check for null values
df.isnull().sum()

# Plot Content Type Distribution
plt.figure(figsize=(6,6))
df['type'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', colors=['#FF9999','#66B3FF'])
plt.title('Content Type Distribution')
plt.ylabel('')
plt.savefig("content_type_distribution.png") # Save the plot as a file
plt.show()

# Plot Content Added per Year
plt.figure(figsize=(10, 6))
df['year_added'].value_counts().sort_index().plot(
    kind='bar', color='skyblue')
plt.title('Content Added per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.savefig("content_per_year.png") # Save the plot as a file
plt.show()

# Plot Top 10 Countries by Content
plt.figure(figsize=(10, 6))
top_countries = df['country'].dropna().str.split(', ').explode().value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title('Top 10 Countries by Content')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.savefig("top_10_countries.png") # Save the plot as a file
plt.show()

from google.colab import files
files.download("content_type_distribution.png")
files.download("content_per_year.png")
files.download("top_10_countries.png")
```
